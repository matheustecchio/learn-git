```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r*-')

plt.subplot(1,2,2)
plt.plot(y,x,'b--')
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('Big')

axes2.plot(y,x)
axes2.set_title('Small')
```

```python
axes1 = plt.figure().add_axes([0.1,0.1,0.8,0.8])
axes1.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
axes[0].set_title('First')

axes[1].plot(y,x)
axes[1].set_title('Second')

plt.tight_layout()
```

```python
fig = plt.figure(figsize=(8,2))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig = plt.figure(figsize=[8,2])
ax = fig.add_axes([0,0,1,1])

ax.plot(x,x**2,label='Squared',color='brown',lw=2,ls='-',
        marker='o',markersize=5)

ax.plot(x,x**3,label='Cubed',color='blue',lw=2,ls='--',
        marker='o',markersize=5)

ax.set_title('Title')
ax.set_xlabel('X')
ax.set_ylabel('Y')

ax.legend()


```
